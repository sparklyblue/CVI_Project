# YOLO Baseline

Goal:
- Detect animals using bounding boxes
- Single-class detection (Animal)
- Benchmark baseline performance before further improvements

Dataset:
- Thermal drone imagery
- Filtered annotations
- Single detection class

```python
from pathlib import Path
import yaml
import torch
from ultralytics import YOLO
import pandas as pd
```

```python
from pathlib import Path
import yaml
import torch
from ultralytics import YOLO

ROOT = Path("..").resolve()
DATASET = ROOT / "dataset_yolo_detection"

print("ROOT:", ROOT)
print("DATASET exists:", DATASET.exists())
print("CUDA:", torch.cuda.is_available())
print("GPU:", torch.cuda.get_device_name(0))
```

```python
runtime_yaml = {
    "path": str(DATASET),
    "train": "images/train",
    "val": "images/val",
    "test": "images/test",
    "nc": 1,
    "names": {0: "Animal"},
}
runtime_yaml_path = ROOT / "runtime_detection.yaml"

with open(runtime_yaml_path, "w") as f:
    yaml.safe_dump(runtime_yaml, f)

print(runtime_yaml_path)
print(runtime_yaml_path.read_text())
```

## Test/Debug Training (v8n)

```python
model = YOLO("yolov8n.pt")

EPOCHS = 3
IMG_SIZE = 1024
BATCH_SIZE = 8
```

```python
results = model.train(
    data=str(runtime_yaml_path),
    epochs=EPOCHS,
    imgsz=IMG_SIZE,
    batch=BATCH_SIZE,
    device=0,
    workers=4,
    project="runs_detection",
    name="baseline_debug",
    pretrained=True,
    verbose=True
)
```

## YOLOv8n Baseline Training

This run trains the first real animal detector baseline after the debug run verified that the dataset, labels, YAML file, and GPU setup work correctly.

```python
model = YOLO("yolov8n.pt")

results = model.train(
    data=str(runtime_yaml_path),
    epochs=20,
    imgsz=1024,
    batch=8,
    device=0,
    workers=4,
    project="runs_detection",
    name="yolov8n_baseline_1024",
    pretrained=True,
    verbose=True
)
```

# YOLOv8s Baseline Training

```python
model = YOLO("yolov8s.pt")

results_s = model.train(
    data=str(runtime_yaml_path),
    epochs=20,
    imgsz=1024,
    batch=4, # safer for heat/VRAM than 8
    device=0,
    workers=4,
    project="runs_detection",
    name="yolov8s_baseline_1024",
    pretrained=True,
    verbose=True
)
```

## Evaluation

```python
ROOT = Path("..").resolve()
runs_dir = ROOT / "runs" / "detect" / "runs_detection"

experiments = {
    "YOLOv8n": runs_dir / "yolov8n_baseline_1024" / "results.csv",
    "YOLOv8s": runs_dir / "yolov8s_baseline_1024" / "results.csv",
}

rows = []

for model_name, csv_path in experiments.items():
    df = pd.read_csv(csv_path)
    df.columns = df.columns.str.strip()

    best_row = df.loc[df["metrics/mAP50(B)"].idxmax()]

    rows.append({
        "Model": model_name,
        "Best epoch": int(best_row["epoch"]),
        "Precision": best_row["metrics/precision(B)"],
        "Recall": best_row["metrics/recall(B)"],
        "mAP50": best_row["metrics/mAP50(B)"],
        "mAP50-95": best_row["metrics/mAP50-95(B)"],
    })

comparison = pd.DataFrame(rows)
comparison
```

## Baseline Training Summary

Two YOLOv8 variants were trained on the wildlife dataset:

- YOLOv8n (Nano)
- YOLOv8s (Small)

The models are compared using precision, recall, mAP@50, and mAP@50-95 on the validation set.

The purpose of this experiment is to determine whether the additional capacity of YOLOv8s produces meaningful performance improvements over YOLOv8n while considering training cost and inference efficiency.
